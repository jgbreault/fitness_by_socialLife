```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("Lab 3 observational study.csv")

data.columns = [
    "ID",
    "a",  # long hair
    "b",  # jeans
    "c",  # warm coat (self)
    "d",  # bright shirt (self)
    "e",  # glasses
    "f",  # height
    "g",  # warm coat (TA)
    "h"   # bright shirt (TA)
]

binary_cols = ["a", "b", "c", "d", "e", "g", "h"]
data_num = data.copy()

for col in binary_cols:
    data_num[col] = data_num[col].map({"Yes": 1, "No": 0})

label_map = {
    "a": "Long hair",
    "b": "Jeans",
    "c": "Warm coat (self)",
    "d": "Bright shirt (self)",
    "e": "Glasses",
    "f": "Height (cm)",
    "g": "Warm coat (TA)",
    "h": "Bright shirt (TA)"
}


cols = ["a", "b", "c", "d", "e", "f", "g", "h"]

corr = data_num[cols].corr()

corr.index = [label_map[c] for c in cols]
corr.columns = [label_map[c] for c in cols]

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heat Map")
plt.show()
```
